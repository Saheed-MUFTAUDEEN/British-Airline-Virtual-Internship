### Data Collection 

In this phase, customers review data was collected from a airline quality website called [skytrax](https://www.airlinequality.com/). Information on the reviews, ratings, date and customer's location was scrapped from this website. 

```python
#Importing libraries 

import pandas as pd 
import requests 
from bs4 import BeautifulSoup 
```

```python
# Create empty list to collect all reviews 
reviews = []

# Create empty list to collect all reviews ratings  
ratings = []

# Create empty list to collect all reviews date
date = []

# Create empty list to collect all the location from which the reviews was made 
location = []
```

```python
# Scrapping info

base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
page = 30
page_size = 100
```

```python
for i in range(1, page + 1):
    print(f"Scrapping page{i}")

    url = f"{base_url}/page{i}/?sortby=post_date%3ADesc&pagesize={page_size}"
    
    response = requests.get(url)
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # get reviews
    for item in soup.find_all("div", {"class": "text_content"}):
        reviews.append(item.get_text())
        
    #get ratings  
    for item in soup.find_all("div", {"class": "rating-10"}):
        try:
            ratings.append(item.span.get_text())
        except:
            print(f"Error on page{i}")
            
    #get the dates  
    for item in soup.find_all("time"):
        date.append(item.get_text())
        
    #get the location
    for item in soup.find_all("h3"):
        location.append(item.span.next_sibling.strip("()"))
        
    print(f" -----------> {len(reviews)} total reviews")
```

```python
ratings = ratings[:3000]
```

```python
# create a dataframe for the scrapped information

df = pd.DataFrame({"Reviews":reviews, "Ratings":ratings, "Date":date, "Location":location})
```
